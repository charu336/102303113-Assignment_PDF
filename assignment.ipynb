```python
import pandas as pd
import numpy as np
df = pd.read_csv("data.csv", encoding="latin1")
# Extract NO2 column (remove missing values)
x = df["no2"].dropna().values
```

```python
a_r = 0.15
b_r = 1.2

z = x + a_r * np.sin(b_r * x)
```

```python
mu = np.mean(z)
```

```python
sigma2 = np.var(z)
lam = 1 / (2 * sigma2)
```

```python
c = np.sqrt(lam / np.pi)
```

```python
print("Estimated Parameters:")
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
